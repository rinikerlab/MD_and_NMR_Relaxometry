```python
import numpy as np
import pandas as pd
import os
import matplotlib as mpl
import matplotlib.pyplot as plt 
from scipy.stats import lognorm
```

```python

fig = plt.figure(layout="constrained", figsize = (9, 9))

axd = fig.subplot_mosaic(
    """
    AB
    CD
    EF
    G.
    """
)
```

```python
def convert_siemons_to_9_state(pop_4_states):
    """
    Convert the data from Lucas Siemons (https://www.rsc.org/suppdata/c9/cc/c9cc06496f/c9cc06496f1.pdf)
    table S5 with same ordering to our data format here.
    """
    pop = np.zeros(9)
    pop[8] = pop_4_states[0]
    pop[5] = pop_4_states[1]
    pop[4] = pop_4_states[2]
    pop[3] = pop_4_states[3]
    return pop * 100
```

```python
def plot_data_to_axis(ax, data_explicit, data_shifts, data_amber, data_charmm, with_leg = False):
    
    rot_names = [r'{$g_p$, $g_p$}', r'{$t$, $g_p$}', r'{$g_m$, $g_p$}',
         r'{$g_p$, $t$}', r'{$t$, $t$}', r'{$g_m$, $t$}',
         r'{$g_p$, $g_m$}', r'{$t$, $g_m$}', r'{$g_m$, $g_m$}',
        ]

    x = np.arange(9) * 1.5
    
    w = 0.2
    if with_leg:
        ax.bar(x-(3/2)*w, data_explicit, width=w, color='black', label = 'Explicit Model')
        ax.bar(x-(1/2)*w, data_shifts, width=w, color='limegreen', label='Chemical Shifts')

        ax.bar(x+(1/2)*w, data_amber, width=w, color='royalblue', label='Anmr-def-3P')
        ax.bar(x+(3/2)*w, data_charmm, width=w, color='purple', label='C36m-def-3P')
    else:
        ax.bar(x-(3/2)*w, data_explicit, width=w, color='black')
        ax.bar(x-(1/2)*w, data_shifts, width=w, color='limegreen')

        ax.bar(x+(1/2)*w, data_amber, width=w, color='royalblue')
        ax.bar(x+(3/2)*w, data_charmm, width=w, color='purple')

    ax.set_xticks(x)
    ax.set_xticklabels(rot_names, fontsize = 7, rotation=25)

    ax.set_ylim([0, 100])

    return None
    
    
```

```python
# Ile 3

x = np.arange(9)

pops_explicit_model = np.array([10, 8, np.nan, 
                                58, 0, 24, 
                                np.nan, np.nan, 24 ])

pops_siemons = convert_siemons_to_9_state([0.078, 0.035, 0.007, 0.881])


pops_amber = np.array([5, 1, 0, 
                       83, 1, 0,  
                       0, 0, 11 ])


pops_charmm = np.array([5, 10, 0, 
                       50, 11, 0, 
                       0, 0, 24,
                       ])

plot_data_to_axis(axd['A'], pops_explicit_model, pops_siemons, pops_amber, pops_charmm)
fig
```
